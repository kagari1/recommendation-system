```python
df[df['Location']=='Er-Rachidia,']='Errachidia,'
df['Location']=df['Location'].str.replace(',','')
location=[]
for i in range(len(df)):
    text=df['Location'][i].split()
    location.append(text[0])


```

```python
df['Location']=location
```

```python
df['Location'].unique()
```

```python
def listToString(s):
    str1=" "
    return (str1.join(s))
```

```python
listToString(df['Location'])
```

```python
from wordcloud import WordCloud
# Instantiate a new wordcloud.
wordcloud = WordCloud(random_state = 8,
 normalize_plurals = False,
 width = 600, height= 300,
 max_words = 300,
 stopwords = [])
# Apply the wordcloud to the text.
wordcloud.generate(listToString(df['Location']))
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1,1, figsize = (9,6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
```
