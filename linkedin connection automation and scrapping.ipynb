```python
job_lists = driver.find_element(By.CLASS_NAME, "scaffold-layout__list-container")
```

```python
links=[]
for page_num in range(1,9):
    y=driver.find_element(By.XPATH,'/html/body/div/div/div/div/div/main/div/section/div/div/ul/li[@data-test-pagination-page-btn=%s]'% (page_num))
    y.click()
    elem=driver.find_element(By.CLASS_NAME,'jobs-search-results-list')
    for i in range(20):
        driver.execute_script('arguments[0].scrollBy(0,250);',elem)
        driver.implicitly_wait(0.2)
    x=job_lists.find_elements(By.XPATH,"//div/div/div/div/div/a[@class='disabled ember-view job-card-container__link job-card-list__title']")
    for i in x:
        links.append(i.get_attribute('href'))
```

```python
job_title = []
job_type=[]
job_location = []
job_description = []

company_name = []
links2= []


for link in links:
    driver.get(link)
    try:
        Title=driver.find_element(By.CLASS_NAME,'jobs-unified-top-card__job-title').text
    except:
        continue
    job_title.append(Title)
    
#     Type=driver.find_element(By.CLASS_NAME,'jobs-unified-top-card__job-insight').text
#     job_type.append(Type)
    
    Name=driver.find_element(By.CLASS_NAME,'jobs-unified-top-card__company-name').text
    company_name.append(Name)
    
#     Location=driver.find_element(By.CLASS_NAME,'jobs-unified-top-card__bullet').text
    Location=driver.find_element(By.XPATH,'/html/body/div/div/div/div/div/div/div/div/div/div/div/span[1]/span[2]').text
    job_location.append(Location)
    
    Description=driver.find_element(By.CLASS_NAME,'jobs-description-content__text').text
    job_description.append(Description.replace('\n',' '))
    
#     Link=driver.find_element(By.CLASS_NAME,'jobs-unified-top-card__job-title').get_attribute('href')
#     links2.append(Link)
    
   
    

    try:
        Link=driver.find_element(By.CLASS_NAME,'jobs-unified-top-card__primary-description').find_element(By.TAG_NAME,'a').get_attribute('href')
    except:
        Link=None
    links2.append(Link)

```

```python
links3=[]
for i in links2:
    links3.append(i.replace("life","about") if i!=None else "")
```

```python
company_infos={"Secteur":"","Presentation":"","Taille de l’entreprise":"","Fondée en":""}
company_infos_list=[]
lks=0
for i in links3:
    lks+=1
    temp_info=company_infos.copy()
    if i=="":
        company_infos_list.append(temp_info)
        continue
    print("Link: %s" % i)
    driver.get(i)
    dt=driver.find_elements(By.XPATH,'/html/body/div/div/div/div/div/div/main/div/div/div/div/section/dl/dt')
    temp_num=0
    for element in dt:
        temp_num+=1
        temp_info[element.text]=driver.find_element(By.XPATH,'/html/body/div/div/div/div/div/div/main/div/div/div/div/section/dl/dd[%s]'% temp_num).text.encode("ascii", "ignore").decode()
    temp_info['Presentation']=driver.find_element(By.CLASS_NAME,'mb5').text
    company_infos_list.append(temp_info)
    print('Links restant %s' % (len(links3)-lks))
```

```python
company_sector=[]
company_presentation=[]
company_size=[]
for comp in company_infos_list:
    company_sector.append(comp['Secteur'])
    company_presentation.append(comp['Presentation'])
    company_size.append(comp['Taille de l’entreprise'])
```

```python
import pandas as pd
```

```python
job_data = pd.DataFrame({
'Company': company_name,
'Title': job_title,
'Location':job_location,
'Description':job_description,
'Sector':company_sector,
'Presentation':company_presentation,
'size':company_size  })

job_data.to_excel('LinkedIn Job Data.xlsx', index = False)
```
